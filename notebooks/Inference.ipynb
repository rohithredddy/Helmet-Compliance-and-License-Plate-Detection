```python
# --------- USER INPUTS ---------
image_path = r"../Inference Data\real imgs\day_18.jpeg"   # <-- Change this to your image path
brightness_factor = 0.2   # <-- Change this value (0.8 darker, 1.5 brighter)
# --------------------------------

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread(image_path)

if img is None:
    print("Image not found. Please check the path.")
else:
    # Convert BGR to RGB for correct display
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Adjust brightness
    bright_img = img.astype(np.float32) * brightness_factor
    bright_img = np.clip(bright_img, 0, 255).astype(np.uint8)
    bright_img_rgb = cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB)

    # Combine images side by side
    combined = np.hstack((img_rgb, bright_img_rgb))

    # Plot combined image
    plt.figure()
    plt.imshow(combined)
    plt.title("             Left: Original                      Right: Brightness Adjusted")
    plt.axis("off")
    plt.show()
```
